```python
#import for arrays, matrices and more
import numpy as np

#import for graph plotting
import matplotlib.pyplot as plt
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15, 8))

# X values on plot.
x = np.linspace(0, 10, 1000)

# Plot polynomial.
ax.plot(x, x**2, label="$x^2$")

# Plot exponential.
ax.plot(x, 2**x, label="$2^x$")

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15,3))

# X values on plot.
x = np.linspace(0, 10, 1000)

# Ploting Lines
ax.plot(x, x, label="$f(x)==g(x)$")
ax.plot(x, x, label="$f(x)==g(x)$")

# Title
ax.set_title('Since both are linear functions, they grow at the same rate, and f(x) is BigO g(x)')

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15,3))

# X values on plot.
x = np.linspace(0, 10, 1000)

# Ploting Lines
ax.plot(x, x**2, label="f(x) = $x^2$")
ax.plot(x, x, label="g(x) = $x$")

# Title
ax.set_title('Since f(x) grows faster than g(x), as x increases f(x) is BigO g(x)')

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15,8))

# X values on plot.
x = np.linspace(0, 10, 1000)

# Ploting Lines
ax.plot(x, x**2 + x, label="f(x) = $x^2+x$")
ax.plot(x, x**2, label="g(x) = $x^2$")

# Title
ax.set_title('f(x) is BigO g(x) because $n^2$ dominates the growth, and $+x$ is almost null comparatively')

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15,5))

# X values on plot.
x = np.linspace(0, 10, 1000)

# Ploting Lines
ax.plot(x, x**3, label="f(x) = $x^3$")
ax.plot(x, x**2, label="g(x) = $x^2$")

# Title
ax.set_title('Since f(x) grows faster than g(x), f(x) is BigO g(x)')

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(15,5))

# X values on plot.
x = np.linspace(1, 10, 1000)

# Ploting Lines
ax.plot(x, np.log(x), label="f(x) = $x log(x)$")
ax.plot(x, x, label="g(x) = $x$")

# Title
ax.set_title('f(x) is BigO because it being a logarithmic term grows slower than the linear g(x)')

# Legend.
ax.legend()

# Show.
plt.show();
```
